```python
df["severidade"] = df["mortos"] * 3 + df["feridos_leves"] * 1
```

```python
uf_perigo = df.groupby("uf")["severidade"].mean().sort_values(ascending=False)

uf_perigo.plot(kind="bar")
plt.title("Severidade mÃ©dia por estado")
plt.ylabel("Severidade")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform ANOVA to compare 'severidade' across 'fase_dia'

# First, group severidade by fase_dia
severidade_por_fase = [df['severidade'][df['fase_dia'] == fase].dropna() for fase in df['fase_dia'].unique()]

f_statistic, p_value = stats.f_oneway(*severidade_por_fase)

print(f"Teste ANOVA para Severidade vs. Fase do Dia:")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("HÃ¡ uma diferenÃ§a estatisticamente significativa na severidade mÃ©dia entre as fases do dia.")
    # Perform Tukey's HSD post-hoc test if ANOVA is significant
    tukey_result = pairwise_tukeyhsd(endog=df['severidade'], groups=df['fase_dia'], alpha=0.05)
    print("\nTeste Post-Hoc de Tukey HSD:")
    print(tukey_result)
else:
    print("NÃ£o hÃ¡ diferenÃ§a estatisticamente significativa na severidade mÃ©dia entre as fases do dia.")


# Calculate confidence intervals for the mean 'severidade' for each 'fase_dia'
conf_intervals = df.groupby('fase_dia')['severidade'].apply(lambda x: stats.t.interval(confidence=0.95, df=len(x)-1, loc=np.mean(x), scale=stats.sem(x)))

print("\nIntervalos de ConfianÃ§a (95%) para a Severidade MÃ©dia por Fase do Dia:")
display(conf_intervals.to_frame(name='Intervalo de ConfianÃ§a'))
```

```python
def anova_tukey(df, target, group):
    grupos = [df[target][df[group] == g] for g in df[group].unique()]
    
    f_stat, p_val = stats.f_oneway(*grupos)

    print(f"\nð ANOVA: {target} vs {group}")
    print(f"F-statistic: {f_stat:.3f}")
    print(f"P-value: {p_val:.5f}")

    if p_val < 0.05:
        print("â¡ï¸ DiferenÃ§a significativa encontrada!")

        tukey = pairwise_tukeyhsd(
            endog=df[target],
            groups=df[group],
            alpha=0.05
        )
        print(tukey)
    else:
        print("â¡ï¸ Sem diferenÃ§a significativa.")
```

```python
categoricas = [
    "fase_dia",
    "tipo_acidente",
    "classificacao_acidente",
    "uf",
    "condicao_metereologica",
    "tipo_pista",
    "tracado_via",
    "uso_solo",
    "sentido_via"
]

for col in categoricas:
    anova_tukey(df, "severidade", col)
```
