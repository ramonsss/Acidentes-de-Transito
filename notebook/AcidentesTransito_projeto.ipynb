```python
causa_impacto = df.groupby('causa_acidente').agg(
    media_pessoas=('pessoas', 'mean'),
    media_mortos=('mortos', 'mean'),
    media_feridos=('feridos', 'mean')
).reset_index()

causa_impacto_mortos = causa_impacto.sort_values(by='media_mortos', ascending=False).head(10)
causa_impacto_feridos = causa_impacto.sort_values(by='media_feridos', ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(y='causa_acidente', x='media_mortos', data=causa_impacto_mortos, palette='Reds_d', hue='causa_acidente', legend=False)
plt.title('Top 10 Causas de Acidentes por MÃ©dia de Mortos')
plt.xlabel('MÃ©dia de Mortos por Acidente')
plt.ylabel('Causa do Acidente')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(y='causa_acidente', x='media_feridos', data=causa_impacto_feridos, palette='Oranges_d', hue='causa_acidente', legend=False)
plt.title('Top 10 Causas de Acidentes por MÃ©dia de Feridos')
plt.xlabel('MÃ©dia de Feridos por Acidente')
plt.ylabel('Causa do Acidente')
plt.tight_layout()
plt.show()
```

```python
df["hora"] = pd.to_datetime(df["horario"]).dt.hour

top_causas = df["causa_acidente"].value_counts().head(10).index
df_top = df[df["causa_acidente"].isin(top_causas)]

causa_hora = df_top.groupby(["hora", "causa_acidente"]).size().unstack().fillna(0)

plt.figure(figsize=(100, 50))

causa_hora.plot()

plt.title("Top 5 causas de acidente por hora")
plt.xlabel("Hora do dia")
plt.ylabel("Quantidade de acidentes")
plt.xticks(range(24))
plt.legend(title="Causa", bbox_to_anchor=(1.05, 1))
plt.grid(alpha=0.3)

plt.show()
```
